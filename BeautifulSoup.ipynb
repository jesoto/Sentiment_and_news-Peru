```python
from bs4 import BeautifulSoup
from urllib.request import urlopen,Request

url_base = 'https://gestion.pe'
```

```python
def capturar_noticias_por_fecha(fecha):
  url=f"{url_base}/archivo/todas/{fecha}/"
  html = urlopen(url)
  soup = BeautifulSoup(html,'html.parser')
  noticias=[]
  for div in soup.find('div',{'role':'main'}).div.find_all('div', recursive=False):
    try:
      a = div.find('a',{'class':"story-item__section text-sm text-black md:mb-15"})
      categoria = a.text
      titulo = div.h2.a.text
      href = div.h2.a['href']
      noticias.append({'categoria':categoria, 'titulo':titulo, 'href':href, 'fecha':fecha})
    except:
      continue
  return noticias

noticias = capturar_noticias_por_fecha("2021-10-02")
noticias
```

```python
def capturar_contenido_de_noticias(noticias):
  for noticia in noticias:
    try:
      href = noticia['href']
      url_noticia=f'{url_base}{href}'
      print(url_noticia)
      html = urlopen(url_noticia)
      soup = BeautifulSoup(html,'html.parser')

      all_paragraphs = soup.section.section.find_all('p') 
      
      text = ' '.join([p.text for p in all_paragraphs])
      noticia['content'] = text
    except:
      continue
  return noticias

noticias = capturar_contenido_de_noticias(noticias)
noticias
```

```python
noticias_gestion=[]
for fecha in ['2021-09-26','2021-09-27','2021-09-28','2021-09-29','2021-09-30','2021-10-01','2021-10-02']:
  noticias = capturar_noticias_por_fecha(fecha)
  noticias = capturar_contenido_de_noticias(noticias)
  noticias_gestion += noticias
noticias_gestion
```
